```python
import requests
import json
import pandas as pd
import numpy as np
import pycountry
from citipy import citipy

# Google developer API key
from api_keys import g_key


#Output Folder Path
path = "output_data/"

#Read_csv_Path
players_top10_uscity_path = f"{path}players_top10_2022_uscity.csv"
players_top10_intcity_path = f"{path}players_top10_2022_intcity.csv"
players_top10_countries_path = f"{path}players_top10_2022_countries.csv"
```

```python
players_top10_uscity = pd.read_csv(players_top10_uscity_path)
players_top10_intcity = pd.read_csv(players_top10_intcity_path)
players_top10_countries = pd.read_csv(players_top10_countries_path)
```

## US City

```python
lats=[]
lngs=[]

# Target city
target_city = players_top10_uscity['US City'].value_counts().index.tolist()

for each, city in enumerate(target_city):
    try:
         # Build the endpoint URL
        target_url = f"https://maps.googleapis.com/maps/api/geocode/json?address={target_city[each]}&key={g_key}"


        # Run a request to endpoint and convert result to json
        geo_data = requests.get(target_url).json()


         # Extract latitude and longitude
        lat = geo_data["results"][0]["geometry"]["location"]["lat" ]
        lng = geo_data["results"][0]["geometry"]["location"]["lng"]

        lats.append(lat)
        lngs.append(lng)

         # Print the latitude and longitude
        print('''
             City: {0}
             Latitude: {1}
             Longitude: {2}
             '''.format(target_city[each], lat, lng))
        
    except Exception as e:
        lats.append("")
        lngs.append("")
        print(f'{target_city[each]} not found or , {e} | Skipping...')
        pass
    
data = {'US City':target_city,'Lat':lats,'Lng':lngs}
players_top10_uscity_gloc = pd.DataFrame(data)

data = {'US City':target_city,'Lat':lats,'Lng':lngs}
players_top10_uscity_gloc = pd.DataFrame(data)

players_top10_uscity_gloc.to_csv(f'{path}players_top10_uscity_gloc.csv', encoding='utf-8-sig')
```

## Countries

```python
lats=[]
lngs=[]

# Target city
target_country = players_top10_countries['Country'].value_counts().index.tolist()
target_country_name = []

for each, country in enumerate(target_country):
    country_conv = pycountry.countries.lookup(target_country[each])
    target_country_name.append(country_conv.name)


for each, city in enumerate(target_country_name):
    try:
         # Build the endpoint URL
        target_url = f"https://maps.googleapis.com/maps/api/geocode/json?address={target_country_name[each]}&key={g_key}"


        # Run a request to endpoint and convert result to json
        geo_data = requests.get(target_url).json()


         # Extract latitude and longitude
        lat = geo_data["results"][0]["geometry"]["location"]["lat" ]
        lng = geo_data["results"][0]["geometry"]["location"]["lng"]

        lats.append(lat)
        lngs.append(lng)

         # Print the latitude and longitude
        print('''
             City: {0}
             Latitude: {1}
             Longitude: {2}
             '''.format(target_country_name[each], lat, lng))
        
    except Exception as e:
        lats.append("")
        lngs.append("")
        print(f'{target_country_name[each]} not found or , {e} | Skipping...')
        pass


data = {'Country':target_country_name,'Lat':lats,'Lng':lngs}
players_top10_countries_gloc = pd.DataFrame(data)

players_top10_countries_gloc.to_csv(f'{path}players_top10_countries_gloc.csv', encoding='utf-8-sig')
```
