```python
import requests
import pandas as pd
import json
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
state = ['AL','AK','AS','AZ','AR','CA','CO','CT','DE','DC','FL',
        'GA','GU','HI','ID','IL','IN','IA','KS','KY','LA','ME',
        'MD','MA','MI','MN','MS','MO','MT','NE','NV','NH','NJ',
        'NM','NY','NC','ND','MP','OH','OK','OR','PA','PR','RI',
        'SC','SD','TN','TX','UT','VT','VA','VI','WA','WV','WI','WY']

# abbr = ['AL','AK']
```

```python
birth_data = []

for abbr in state:
    print(abbr)
    
    url = f'https://www.basketball-reference.com/friv/birthplaces.fcgi?country=US&state={abbr}'
    html = urlopen(url)
    soup = BeautifulSoup(html, features='lxml')
    
    try:
        headers = [th.getText() for th in soup.findAll('tr', limit=2)[1].findAll('th')]
    except Exception as e:
        print(e)
        pass
    try:
        rows = soup.findAll('tr')[2:]
    except Exception as e:
        print(e)
        pass
    try:
        rows_data = [[td.getText() for td in rows[i].findAll('td')]for i in range(len(rows))]
    except Exception as e:
        print(e)
        pass
        
    last_year = 2022
    for i in range(0, len(rows_data)):
        rows_data[i].insert(0, last_year)
        last_year -=1
        
    df = pd.DataFrame(rows_data, columns = headers)
    df['State'] = abbr
    
    birth_data.append(df)
    
birth_data = pd.concat(birth_data)
```

```python
cols = ['Player','Yrs','From','To','Date','City','State']
birth_data = birth_data.loc[:,cols]
birth_data
```

```python
country = ['AO','AG','AR','AU','AT','BS','BE','BA','BR','BG','CM','CA','CV','CN','CO',
        'HR','CU','CZ','CD','DK','DM','DO','EG','EE','FI','FR','GF','GA','GE','DE',
        'GH','GR','GP','GN','GY','HT','HU','IS','IE','IR','IL','IT','JM','JP','LV','LB',
        'LT','LU','ML','MQ','MX','ME','MA','NL','NZ','NG','NO','PA','PL','PT','PR','KR','MK',
        'CG','RO','RU','LC','VC','SN','RS','SK','SI','ZA','SS','ES','SD','SE','CH','TW',
        'TT','TN','TR','VI','UA','GB','TZ','UY','VE']

ca_state = ['MB','NS','ON','QC','SK','BC']
```

```python
def get_player_info(url, co, co_birth_data):
    print(co)
    headers = []
    rows_data = []

    html = urlopen(url)
    soup = BeautifulSoup(html, features='lxml')
    
    try:
        headers = [th.getText() for th in soup.findAll('tr', limit=2)[1].findAll('th')]
        rows = soup.findAll('tr')[2:]
        rows_data = [[td.getText() for td in rows[i].findAll('td')]for i in range(len(rows))]
    except Exception as e:
        print(e)
        pass
        
    last_year = 2022
    for i in range(0, len(rows_data)):
        rows_data[i].insert(0, last_year)
        last_year -=1
        
    df = pd.DataFrame(rows_data, columns = headers)
    df['Country'] = co
    
    co_birth_data = co_birth_data.append(df)
    return co_birth_data
```

```python
co_birth_data = pd.DataFrame()

for co in country:
    
    if co == 'CA':
        for st in ca_state:
            url = f'https://www.basketball-reference.com/friv/birthplaces.fcgi?country={co}&state={st}'
            co_birth_data = get_player_info(url, co, co_birth_data)
    else:
        url = f'https://www.basketball-reference.com/friv/birthplaces.fcgi?country={co}&state='
        co_birth_data = get_player_info(url, co, co_birth_data)
```
