```python
def get_anthro_df(response):
    
    data_dict = response.get_dict()

    results = data_dict['resultSets'][0]

    header = results['headers']

    anthro = results['rowSet']

    anthro = pd.DataFrame(anthro, columns=header)

    return anthro
```

```python
from nba_api.stats.endpoints import draftcombineplayeranthro

seasons = ['2002-03', '2003-04', '2004-05', '2005-06', '2006-07', '2007-08', '2008-09', '2009-10', 
           '2010-11', '2011-12', '2012-13', '2013-14', '2014-15', '2015-16', '2016-17', '2017-18',
           '2018-19', '2019-20', '2020-21', '2021-22']
player_df = []

for season in seasons:
    print(season)
    
    try: 
        response = draftcombineplayeranthro.DraftCombinePlayerAnthro(league_id='00',season_year=seasons)
        dictionary = response.get_dict()
        result = dictionary['resultSets'][0]
        header = result['headers']
        data = result['rowSet']
        df = pd.DataFrame(data, columns=header)
        player_df.append(df)
    except Exception as e:
        print (e)
        pass

# print(header)
# pp.pprint(data)
player_df = pd.concat(player_df)
player_df
```

```python
cols = ['PLAYER_ID', 'PLAYER_NAME', 'POSITION', 'WINGSPAN', 'STANDING_REACH']
anthro_df = player_df.loc[:,cols]
anthro_df
```

```python
anthro_df.to_csv('output/anthro.csv')
```
